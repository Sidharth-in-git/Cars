```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Cars.csv", encoding='latin1')
df.head(40)
```

```python
df.columns
```

```python
plt.hist(df['Company Names'].head(40), bins=20, edgecolor='black')
plt.tight_layout()
plt.xticks(rotation=90, fontsize=10)
plt.xlabel('BRANDS', fontsize=20, font='Times New Roman')
plt.show()

```

```python
plt.figure(figsize=(8,8))
plt.bar(df.head(20)['Company Names'],df.head(20)['Cars Names'],color='green', edgecolor='black')
plt.xlabel('COMPANIES', font='Times New Roman', fontsize='15')
plt.ylabel('NAMES', font='Times New Roman', fontsize=15 )
plt.xticks(fontsize='11', rotation=90)
plt.grid(color='black')

plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(df['Cars Names'].head(20), df['Engines'].head(20), marker="o", color='green')
plt.xticks(rotation=90, fontsize=10)
plt.grid(True)
plt.xlabel('NAMES')
plt.ylabel('Engines')
plt.show()
```
